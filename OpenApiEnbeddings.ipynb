```python
import requests
import pandas as pd
import openai
import yfinance as yf
from datetime import datetime, timedelta

# Set your API keys
FMP_API_KEY = ''
openai.api_key = ''
```

```python
def get_aapl_news(days=7):
    url = f"https://financialmodelingprep.com/api/v3/stock_news?tickers=AAPL&limit=100&apikey={FMP_API_KEY}"
    response = requests.get(url)
    news = pd.DataFrame(response.json())
    news['publishedDate'] = pd.to_datetime(news['publishedDate'])
    news = news[news['publishedDate'] >= datetime.now() - timedelta(days=days)]
    return news[['title', 'publishedDate']]
    
news_df = get_aapl_news()
news_df.head()
```

```python
def get_embedding(text):
    result = openai.embeddings.create(
        input=[text],
        model="text-embedding-3-small"
    )
    return result['data'][0]['embedding']

def get_sentiment(text):
    system_prompt = "Rate this financial news from -1 (very bad) to 1 (very good) for Apple stock:"
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": text}
        ]
    )
    score = float(response['choices'][0]['message']['content'].strip())
    return score
```
